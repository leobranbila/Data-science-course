```python
#Gráficos de dispersão
plt.scatter(x=df.Girth,y=df.Volume, color='blue', facecolors = 'none', marker='*')
plt.title('Árvores')
plt.xlabel('Circuferência')
plt.ylabel('Volume')
```

```python
plt.plot(df.Girth,df.Volume)
plt.title('Árvores')
plt.xlabel('Circuferência')
plt.ylabel('Volume')
```

```python
#Gráfico de dispersão com "Afastamento dos dados"(jitter). Que seria quando um gráfico
# possa ter mais de um registro igual ou muito parecido então para não ficar pontos sobrepostos
#  usamos o parâmetro jitter para gerar um pequeno valor aleatório fazendo com que esses dados tenham um pequeno distanciamento
#fit_reg = linha de tendência
sns.regplot(x=df.Girth,y=df.Volume, data = df, x_jitter=0.3, fit_reg=False)
```

```python
#Repetindo comando para ver um pouco mais do jiiter em ação com diferentes valores 
sns.regplot(x=df.Girth,y=df.Volume, data = df, x_jitter=0.5, fit_reg=False)
```

```python
x = df.conc
y = df.uptake
```

```python
#Gerará uma lista com apenas valores UNICOS trazidos de uma coluna
unicos = list(set(df.Treatment))
unicos
```

```python
#Percorre cada tipo de tratamento (da lista criada) e cria um gráfico de dispersão
for i in range(len(unicos)):
    indice = df.Treatment == unicos[i]
    plt.scatter(x[indice], y[indice], label = unicos[i])
plt.legend(loc = 'lower right')
```
